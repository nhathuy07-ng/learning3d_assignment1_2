```python
file = open("data/rgbd_data.pkl", "rb")
pkl_file = pickle.load(file)
file.close()
```

```python
points = torch.Tensor(pkl_file['rgb2'])
rgb = torch.Tensor(pkl_file['rgb1'])
point_cloud = pytorch3d.structures.Pointclouds(points=points, features=rgb)
```
